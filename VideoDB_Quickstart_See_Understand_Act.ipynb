## 🧠 UNDERSTAND: Index & Search

To unlock semantic search, we need to **index** the video. VideoDB offers two types of indexes:

1. **🗣️ Spoken Words**: Transcribes audio and makes speech searchable
2. **🖼️ Visual Scenes**: Analyzes frames to find objects, formulas, and events

Let's create both!

---

### 🗣️ Index Spoken Words

This generates a word-level transcript with precise timestamps:

```python
# Index spoken content
video.index_spoken_words()
print("✅ Spoken word index complete!")
```

Let's look at the transcript. Each word has a start and end timestamp:

```python
# Get word-level transcript
transcript = video.get_transcript()

print(f"Total words: {len(transcript)}\n")
print("First 10 words:")
for entry in transcript[:10]:
    print(f"  [{entry['start']:.1f}s - {entry['end']:.1f}s] {entry['text']}")
```

Let's look at the entire transcript.

```python
transcript_text = video.get_transcript_text()
transcript_text
```

### 🔍 Search Spoken Words

Now we can search for any keyword. Let's find all mentions of "transformer":

```python
from videodb import SearchType, IndexType

# Define the search query
search_query = "transformer"

# Search for "transformer" in spoken content
results = video.search(
    query=search_query,
    search_type=SearchType.keyword,
    index_type=IndexType.spoken_word
)
print(f"Search results for query {search_query}:\n")
print(f"Total mentions found: {len(results.shots)}\n")

for i, shot in enumerate(results.shots):
    start_minutes, start_seconds = divmod(shot.start, 60)
    end_minutes, end_seconds = divmod(shot.end, 60)

    print(f"--- Mention {i+1} ---")
    print(f"  Time: {int(start_minutes):02d}:{start_seconds:05.2f} - {int(end_minutes):02d}:{end_seconds:05.2f}")
    print("-" * 30)
```

Lets compile and stream the results

```python
# Compile results into a playable stream
stream_url = results.compile()
print(f"Search results compiled: {stream_url}")

# Play the highlights
play_stream(stream_url)
```

### 🔍 Search Visual Content

Now we can search for visual elements. Let's find scenes with mathematical formulas or matrices:

```python
# Search for mathematical formulas in visual content
visual_search_query = "mathematical formula or matrix"
visual_results = video.search(
    query=visual_search_query,
    search_type=SearchType.semantic,
    index_type=IndexType.scene
)

print(f"Visual search results for query {visual_search_query}:\n")
print(f"Total results found: {len(visual_results.shots)}\n")

print("-" * 100)

for i, shot in enumerate(visual_results.shots):
    start_minutes, start_seconds = divmod(shot.start, 60)
    end_minutes, end_seconds = divmod(shot.end, 60)

    print(f"\n--- Visual Mention {i+1} ---")
    print(f"\nTime: {int(start_minutes):02d}:{start_seconds:05.2f} - {int(end_minutes):02d}:{end_seconds:05.2f}")
    if hasattr(shot, 'text'):
        print(f"\nDescription: {shot.text}\n")
    print("-" * 100)
```

### 📝 Prepare Transcript for AI Analysis

First, we'll merge the word-level transcript into sentences for easier analysis:

```python
import json

def merge_transcript_to_sentences(word_transcript):
    """Merge word-level transcript into sentence-level"""
    sentence_transcript = []
    current_sentence_words = []

    for i, entry in enumerate(word_transcript):
        text = entry['text']

        if text == '-' and len(current_sentence_words) > 0:
            first_word = current_sentence_words[0]
            last_word = current_sentence_words[-1]
            sentence_text = ' '.join(w['text'] for w in current_sentence_words if w['text'] != '-')
            sentence_text = sentence_text.strip()

            if sentence_text:
                sentence_transcript.append({
                    'start': first_word['start'],
                    'end': last_word['end'],
                    'text': sentence_text
                })
            current_sentence_words = []
        elif text != '-':
            current_sentence_words.append(entry)

        if i == len(word_transcript) - 1 and len(current_sentence_words) > 0:
            first_word = current_sentence_words[0]
            last_word = current_sentence_words[-1]
            sentence_text = ' '.join(w['text'] for w in current_sentence_words if w['text'] != '-')
            sentence_text = sentence_text.strip()

            if sentence_text:
                sentence_transcript.append({
                    'start': first_word['start'],
                    'end': last_word['end'],
                    'text': sentence_text
                })
    return sentence_transcript

# Generate sentence-level transcript
sentence_transcript = merge_transcript_to_sentences(transcript)

# Compact format for AI analysis
compact_transcript = [
    {"s": round(e["start"], 2), "e": round(e["end"], 2), "t": e["text"]}
    for e in sentence_transcript
]

transcript_str = json.dumps(compact_transcript)
print(f"✅ Prepared {len(compact_transcript)} sentences ({len(transcript_str)} chars) for AI analysis")
print(f"Compact Transcript: {json.dumps(compact_transcript[:5], indent=2)}")
```

### 🤖 Find Best Moment with AI

Now let's ask an AI to analyze the transcript and identify the single best educational moment (60-90 seconds):

```python
# Prompt for finding the best moment
find_moment_prompt = f"""You are analyzing an ML lecture transcript to find the best standalone educational moment.

Identify ONE segment (60-90 seconds) that:
- Explains a key concept completely
- Works as a standalone clip
- Has clear teaching value
- Begins and ends at natural boundaries

TRANSCRIPT (s=start seconds, e=end seconds, t=text):
{transcript_str}

Return ONLY valid JSON:
{{
  "title": "<short descriptive title>",
  "start": <exact start timestamp>,
  "end": <exact end timestamp>,
  "description": "<why this is a great educational moment>"
}}

RULES:
- Use exact 's' and 'e' values from transcript
- Duration must be 60-90 seconds
- Return ONLY the JSON, no extra text
"""

# Ask AI to find the best moment
response = coll.generate_text(
    prompt=find_moment_prompt,
    model_name="pro",
    response_type="json"
)

best_moment = response["output"]

print(f"🎯 Best Moment Found!")
print(f"Title: {best_moment['title']}")
print(f"Time: {best_moment['start']:.1f}s → {best_moment['end']:.1f}s ({best_moment['end'] - best_moment['start']:.0f}s)")
print(f"Why: {best_moment['description']}")
```

### 🎬 Create Edited Video with Timeline Editor

Now we'll compose everything into a polished video using VideoDB's Timeline Editor. Think of it like layers in Photoshop:

- **Timeline**: The canvas
- **Track**: One layer (video, image, text, audio)
- **Clip**: A positioned element in a track
- **Asset**: The media (video, image, audio, text)

Let's build it!

```python
import textwrap
from videodb.editor import (
    ImageAsset, TextAsset,
    Font, Background, Border, Shadow,
    TextAlignment, Position, Offset,
    CaptionAsset, FontStyling, BorderAndShadow,
    Positioning, CaptionAnimation, CaptionAlignment
)

# Image assets
bg_image_asset = ImageAsset(id=bg_image.id)
logo_asset = ImageAsset(id=logo.id)

# Caption asset
caption_asset = CaptionAsset(
    src="auto",
    font=FontStyling(name="Clear Sans", size=42, bold=True),
    primary_color="&H00FFFFFF",      # White text
    secondary_color="&H00BB8800",     # Dark Cyan highlight
    border=BorderAndShadow(outline=2, outline_color="&H000E0D0C"),  # Dark outline
    position=Positioning(
        alignment=CaptionAlignment.bottom_center,
        margin_v=40,
    ),
    animation=CaptionAnimation.box_highlight,
)

# Helper: auto-sized text box with wrapping
def make_text(text, size=60, max_chars=30):
    lines = textwrap.wrap(text, width=max_chars)
    wrapped = "\n".join(lines)
    longest = max(len(l) for l in lines)
    w = int(longest * size * 0.65 + 40)
    h = int(len(lines) * size * 1.3 + 30)
    return TextAsset(
        text=wrapped,
        font=Font(family="Clear Sans", size=size, color="#FFFFFF", opacity=1.0),  # White text
        border=Border(color="#0088BB", width=3.0),  # Dark Cyan border
        shadow=Shadow(color="#000000", x=3, y=3),
    )

clip_title=best_moment['title']
intro_text_asset = make_text(clip_title)
outro_text_asset = make_text("Learn More at VideoDB.io")

# Durations
intro_duration = 2
outro_duration = 2

print(f"Intro/Outro: {intro_duration}s each")
print(f"Main clip  : {clip_duration:.0f}s")
print(f"Total      : {intro_duration + clip_duration + outro_duration:.0f}s")
```

Now let's build the timeline with all layers:

```python
from videodb.editor import Timeline, Track, Clip, VideoAsset

timeline = Timeline(conn)
timeline.resolution = "1280x720"

# --- Intro track (background image) ---
intro_track = Track()
intro_track.add_clip(0, Clip(asset=bg_image_asset, duration=intro_duration))
timeline.add_track(intro_track)

# --- Intro text track ---
intro_text_track = Track()
intro_text_track.add_clip(0, Clip(
    asset=intro_text_asset,
    duration=intro_duration,
    position=Position.center,
    offset=Offset(x=0, y=0.2),
))
timeline.add_track(intro_text_track)

# --- Main video track with captions (same track, same start time) ---
main_track = Track()
main_track.add_clip(intro_duration, Clip(
    asset=VideoAsset(id=video.id, start=clip_start),
    duration=clip_duration,
))
main_track.add_clip(intro_duration, Clip(
    asset=caption_asset,
    duration=clip_duration,
))
timeline.add_track(main_track)

# --- Logo overlay track (during main video) ---
logo_track = Track()
logo_track.add_clip(intro_duration, Clip(
    asset=logo_asset,
    duration=clip_duration,
    fit=None,
    scale=0.125,
    opacity=0.5,
    position=Position.top_right,
    offset=Offset(x=-0.02, y=0.02),
))
timeline.add_track(logo_track)

# --- Outro track (background image) ---
outro_start = intro_duration + clip_duration

outro_track = Track()
outro_track.add_clip(outro_start, Clip(asset=bg_image_asset, duration=outro_duration))
timeline.add_track(outro_track)

# --- Outro text track ---
outro_text_track = Track()
outro_text_track.add_clip(outro_start, Clip(
    asset=outro_text_asset,
    duration=outro_duration,
    position=Position.center,
    offset=Offset(x=0, y=0.2),
))
timeline.add_track(outro_text_track)

print("✅ Timeline ready with intro, main video + captions, logo, and outro")
```

## 🎬 ACT: Create Events & Alerts

Now let's set up automated alerts. When the AI detects an intruder, it will send a web-socket notification.

---

### 🚨 Create Detection Event

```python
# Define what we want to detect
event_id = conn.create_event(
    event_prompt="Detect if a person is present in the monitored area",
    label="intruder_detected",
)

print(f"✅ Event created")
print(f"   Event ID: {event_id}")
print(f"   Label: intruder_detected")
```

### 🔌 Connect WebSocket for Real-time Delivery

WebSockets let us receive alerts instantly as they happen. We'll connect to VideoDB's WebSocket server:

```python
import asyncio
import json

# Connect to WebSocket
ws_wrapper = conn.connect_websocket()
ws = await ws_wrapper.connect()

print(f"✅ WebSocket connected")
print(f"   Connection ID: {ws.connection_id}")
```

### 🔔 Create Alert with WebSocket Delivery

Now we attach the alert to our scene index. Every time the event is detected, we'll receive a real-time notification with a video clip:

```python
# Create alert with WebSocket delivery
alert_id = scene_index.create_alert(
    event_id=event_id,
    callback_url="",  # Empty = WebSocket only
    ws_connection_id=ws.connection_id
)

print(f"🔔 Alert created!")
print(f"   Alert ID: {alert_id}")
print(f"   Delivery: Real-time WebSocket")
```

### 👂 Listen for Intruder Alerts

Let's listen for 30 seconds. When an intruder is detected, we'll automatically receive and display the alert with the video clip:

```python
# Store alerts for later replay
received_alerts = []

# Listen for alerts in real-time
async def listen_for_alerts():
    timeout = 30
    print(f"⏱️ Listening for alerts ({timeout} seconds)...")
    print("💡 The security stream will trigger alerts when intruders are detected\n")

    try:
        async with asyncio.timeout(timeout):
            async for msg in ws.receive():
                # Only process alert channel messages
                if msg.get("channel") == "alert":
                    received_alerts.append(msg)
                    alert_num = len(received_alerts)

                    data = msg.get("data", {})

                    print(f"\n🚨 ALERT #{alert_num} RECEIVED!")
                    print(f"   Event: {data.get('label', 'N/A')}")
                    print(f"   Time: {msg.get('timestamp', 'N/A')}")
                    print(f"   Confidence: {data.get('confidence', 'N/A')}")
                    print(f"   📹 Clip URL: {data.get('stream_url', 'N/A')}")
                    print(f"   💡 Explanation: {data.get('explanation', 'N/A')}")

    except asyncio.TimeoutError:
        print(f"\n✅ Listening complete ({len(received_alerts)} alert(s) received)")

# Start listening
await listen_for_alerts()
```

### 📺 Preview Alert Clips

Let's replay one of the alert clips we received. Each alert includes a video segment showing the exact moment the intruder was detected:

```python
# Play the first alert clip

first_alert = received_alerts[0]
clip_url = first_alert.get("data", {}).get("stream_url")

print(f"Playing Alert #1:")
print(f"   Event: {first_alert['data']['label']}")
print(f"   Explanation: {first_alert['data']['explanation']}\n")
play_stream(clip_url)
```

#### Lets close the web socket after use.

```python
ws_wrapper.close()
```

**💡 What just happened?**

- We connected to VideoDB's WebSocket for real-time event delivery

- Each alert includes a video clip showing the exact moment of detection

**⚡ Why WebSockets?**
- **Instant delivery**: Alerts arrive in real-time as events happen
- **Live dashboards**: Perfect for building interactive monitoring interfaces
- **Two-way communication**: Can send and receive data on the same connection
